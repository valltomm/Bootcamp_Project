```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import gmaps

# Google developer API key
#from gkey import gkey

# Access maps with unique API key
#gmaps.configure(api_key=gkey)
```

```python
homicide_path= "Homicide.csv"
```

```python
homicide_rawdata = pd.read_csv(homicide_path)
```

```python
homicide_rawdata.head()
```

```python
homicide_data.dtypes
```

```python
#select relevant columns
homicide_data = homicide_rawdata.loc[:, ['Index_','Occurrence_year',"Lat","Long",'Neighbourhood']]
#rename column names 
homicide_data .rename(columns={'Index_':'Homicide_Index',
                          'Occurrence_year':'Homicide_Year',
                          'Lat':'Homicide_Lat',
                          'Long':'Homicide_Lng',
                          'Neighbourhood':'Homicide_Neighbourhood'},inplace=True)


# !!! vt: property data for 2016

#select only 2015 record   !! vt: changed to 2016 
homicide_data = homicide_data.loc[homicide_data["Homicide_Year"]==2016]

homicide_data.head()
```

```python
#check if any missing values and found none
homicide_data.count()
```

```python

# !!! missing d in locations_homicie    revised to locations_homicide

locations_homicide = homicide_data[["Homicide_Lat", "Homicide_Lng"]]
print(locations_homicide)
```

```python
figure_layout = {
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Assign the marker layer to a variable
markers = gmaps.marker_layer(locations_homicide)
# Add the layer to the map
fig.add_layer(markers)
fig
```

```python
#adding heat layer  
heat_layer = gmaps.heatmap_layer(locations_homicide, 
                                 dissipating=False, max_intensity=10)


fig.add_layer(heat_layer)

fig
```
